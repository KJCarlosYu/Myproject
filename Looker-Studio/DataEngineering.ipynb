# Replacing the abbreviation of countries to their full name

## don't change this part

```python
url = "https://ec.europa.eu/eurostat/statistics-explained/index.php?title=Glossary:Country_codes"
tables = pd.read_html(url)
```

```python
tables[0]
```

```python
country_codes_df = tables[0]

country_names = pd.concat([country_codes_df.iloc[:, 0], country_codes_df.iloc[:, 2], country_codes_df.iloc[:, 4], country_codes_df.iloc[:, 6]])
country_codes = pd.concat([country_codes_df.iloc[:, 1], country_codes_df.iloc[:, 3], country_codes_df.iloc[:, 5], country_codes_df.iloc[:, 7]])

new_country_codes_df = pd.DataFrame({'Country name': country_names.dropna().reset_index(drop=True), 'Country code': country_codes.dropna().reset_index(drop=True)})

print(new_country_codes_df)
```

```python
new_country_codes_df
```

```python
new_country_codes_df['Country code'] = new_country_codes_df['Country code'].str.extract(r'(\w+)')
```
